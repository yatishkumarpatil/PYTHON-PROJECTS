1.What is the average median income of the data set and check the distribution of data using appropriate plots. Please explain the distribution of the plot.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Loading housing dataset into pandas
df = pd.read_excel("housing.xlsx")

# Calculating the average median income
avg_median_income = df['median_income'].mean()
print(f"Average median income is {avg_median_income}")

# histogram of the median_income
plt.hist(df['median_income'], bins=30)
plt.xlabel('Median Income')
plt.ylabel('Frequency')
plt.title('Bar Plot Median Income')
plt.show()
```

The feature median_income is a continuous feature.The histogram of the median income is showing a skewness towards right. which means that majority of the median income values are on left side and are low income.there are a few high income median values which are outliers.

5.Create a data set by filling the missing data with the mean value of the total_bedrooms in the original data set.

```python
from sklearn.impute import SimpleImputer

# Loading housing dataset into pandas
df = pd.read_excel("housing.xlsx")

# Creating an instance of the SimpleImputer class
imputer = SimpleImputer(strategy="mean")

# fitting the imputer to the data set
imputer.fit(df[["total_bedrooms"]])

# filling the missing values in total_bedrooms with the mean value
new_df = df.copy()
new_df["total_bedrooms"] = imputer.transform(new_df[["total_bedrooms"]])

# creating a dataset with filled missing values
new_df.to_excel("fill_miss.xlsx")
```

```python
# Loading housing dataset into pandas
df = pd.read_excel("housing.xlsx")

# checking for any null values in the housing dataset
print(df.isnull().sum())
```

```python
# loading the filled dataset
new_df = pd.read_excel("fill_miss.xlsx")

# checking for null values in filled dataset
print(new_df.isnull().sum())
```

6.Write a programming construct (create a user defined function) to calculate the median value of the data set wherever required.

```python
# Loading housing dataset into pandas
df = pd.read_excel("housing.xlsx")

# function to calculates the median value of the data set 
def median_value(df, column):
    return df[column].median()

median_value = median_value(df, 'population')

print(f"The population median is {median_value}")
```

10.Please create a new column named total_bedroom_size. If the total bedrooms is 10 or less, it should be quoted as small. If the total bedrooms is 11 or more but less than 1000, it should be medium, otherwise it should be considered large.

```python
# Loading housing dataset into pandas
df = pd.read_excel("housing.xlsx")

# Creating new column total_bedroom_size
df['total_bedroom_size'] = df['total_bedrooms'].apply(lambda x: 'small' if x <= 10 else 'medium' if x < 1000 else 'large')

df.head()
```
